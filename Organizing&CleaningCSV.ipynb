```python
import pandas as pd
import numpy as np

Uploaded_data = pd.read_csv("environment_station_data_updated.csv")

STANDARD_COLUMNS = ['Station_ID', 'Scientific_name', 'Locality', 'Temperature','Latitude', 'Longitude', 'Salinity/PSU', 'Dissolved_O2_mgL','Current_m_s', 'Collection_Date']

def normalize_column(col_name):

    norm = col_name.replace("_","").replace(" ","").replace("(","").replace(")","").lower()
    if norm.endswith('c'):
        norm = norm[:-1]
    return norm

col_map = {}
for std_col in STANDARD_COLUMNS:
    norm_std = normalize_column(std_col)
    for uploaded_col in Uploaded_data.columns:
        norm_uploaded = normalize_column(uploaded_col)
        if norm_std in norm_uploaded or norm_uploaded in norm_std:
            col_map[uploaded_col] = std_col

existing_cols = list(col_map.keys())
Standardized_format = Uploaded_data[existing_cols].copy()
Standardized_format.rename(columns=col_map, inplace=True)

Standardized_format = Standardized_format.drop_duplicates()

if 'Station_ID' in Standardized_format.columns:
    Standardized_format['Station_ID'] = Standardized_format['Station_ID'].astype(str)
    pattern_Station_ID = r"^SS\d{5}$"
    Standardized_format["Station_ID"] = Standardized_format["Station_ID"].where(Standardized_format["Station_ID"].str.match(pattern_Station_ID), 'null')

if 'Scientific_name' in Standardized_format.columns:
    Standardized_format['Scientific_name'] = Standardized_format['Scientific_name'].astype(str)

if 'Locality' in Standardized_format.columns:
    Standardized_format['Locality'] = Standardized_format['Locality'].astype(str)

if 'Temperature' in Standardized_format.columns:
    Standardized_format['Temperature'] = pd.to_numeric(Standardized_format["Temperature"], errors="coerce")
    min_temp = -10
    max_temp = 60
    Standardized_format['Temperature'] = Standardized_format['Temperature'].where((Standardized_format['Temperature'] >= min_temp) & (Standardized_format['Temperature'] <= max_temp), '-')

if 'Latitude' in Standardized_format.columns:
    Standardized_format['Latitude'] = pd.to_numeric(Standardized_format["Latitude"], errors="coerce")

if 'Longitude' in Standardized_format.columns:
    Standardized_format['Longitude'] = pd.to_numeric(Standardized_format["Longitude"], errors="coerce")

if 'Salinity/PSU' in Standardized_format.columns:
    Standardized_format['Salinity/PSU'] = pd.to_numeric(Standardized_format["Salinity/PSU"], errors="coerce")

if 'Dissolved_O2_mgL' in Standardized_format.columns:
    Standardized_format['Dissolved_O2_mgL'] = pd.to_numeric(Standardized_format["Dissolved_O2_mgL"], errors="coerce")

if 'Current_m_s' in Standardized_format.columns:
    Standardized_format['Current_m_s'] = pd.to_numeric(Standardized_format["Current_m_s"], errors="coerce")

if 'Collection_Date' in Standardized_format.columns:
    Standardized_format['Collection_Date'] = pd.to_datetime(Standardized_format['Collection_Date'], errors='coerce')
    Standardized_format['Collection_Date'] = Standardized_format['Collection_Date'].dt.strftime('%d-%m-%Y')

Standard_taxonomy_col = ['Station_ID', 'Species_ID', 'Scientific_name', 'Common_name', 'Phylum', 'Class', 'Order', 'Family', 'Genus', 'Species', 'Abundance']

def normalize_column(col_name):
    norm = col_name.replace("_","").replace(" ","").replace("(","").replace(")","").lower()
    return norm

col_map = {}
for std_col in Standard_taxonomy_col:
    norm_std = normalize_column(std_col)
    for uploaded_col in Uploaded_data.columns:
        norm_uploaded = normalize_column(uploaded_col)
        if norm_std in norm_uploaded or norm_uploaded in norm_std:
            col_map[uploaded_col] = std_col

existing_cols = list(col_map.keys())
Standard_taxonomy_tab = Uploaded_data[existing_cols].copy()
Standard_taxonomy_tab.rename(columns=col_map, inplace=True)

Standard_taxonomy_tab = Standard_taxonomy_tab.drop_duplicates()

if 'Station_ID' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Station_ID'] = Standard_taxonomy_tab['Station_ID'].astype(str)
    pattern_Station_ID = r"^SS\d{5}$"
    Standard_taxonomy_tab["Station_ID"] = Standard_taxonomy_tab["Station_ID"].where(Standard_taxonomy_tab["Station_ID"].str.match(pattern_Station_ID), '-')

if 'Species_ID' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Species_ID'] = Standard_taxonomy_tab['Species_ID'].astype(str)
    pattern_Station_ID = r"^SP-\d{3}$"  #here sp should  also be added that logic is to be added
    Standard_taxonomy_tab["Species_ID"] = Standard_taxonomy_tab["Species_ID"].where(Standard_taxonomy_tab["Species_ID"].str.match(pattern_Station_ID), '-')

if 'Scientific_name' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Scientific_name'] = Standard_taxonomy_tab['Scientific_name'].astype(str)

if 'Common_name' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Common_name'] = Standard_taxonomy_tab['Common_name'].astype(str)

if 'Phylum' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Phylum'] = Standard_taxonomy_tab['Phylum'].astype(str)

if 'Class' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Class'] = Standard_taxonomy_tab['Class'].astype(str)
    
if 'Order' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Order'] = Standard_taxonomy_tab['Order'].astype(str)
    
if 'Family' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Family'] = Standard_taxonomy_tab['Family'].astype(str)
    
if 'Genus' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Genus'] = Standard_taxonomy_tab['Genus'].astype(str)

if 'Species' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Species'] = Standard_taxonomy_tab['Species'].astype(str)
    
if 'Abundance' in Standard_taxonomy_tab.columns:
    Standard_taxonomy_tab['Abundance'] = pd.to_numeric(Standard_taxonomy_tab["Abundance"], errors="coerce")

STANDARD_eDNA_COLUMNS = ['Station_ID', 'Sample_ID', 'Scientific_name', 'Sequence_ID', 'eDNA_ReadCount', 'Marker_Gene', 'Relative_Abundance_Percent', 'eDNA_Collection_Date']

def normalize_column(col_name):
    norm = col_name.replace("_","").replace(" ","").replace("(","").replace(")","").lower()
    return norm

col_map = {}
for std_col in STANDARD_eDNA_COLUMNS:
    norm_std = normalize_column(std_col)
    for uploaded_col in Uploaded_data.columns:
        norm_uploaded = normalize_column(uploaded_col)
        if norm_std in norm_uploaded or norm_uploaded in norm_std:
            col_map[uploaded_col] = std_col

existing_cols = list(col_map.keys())
Standardized_eDNA = Uploaded_data[existing_cols].copy()
Standardized_eDNA.rename(columns=col_map, inplace=True)

Standardized_eDNA = Standardized_eDNA.drop_duplicates()

if 'Station_ID' in Standardized_eDNA.columns:
    Standardized_eDNA['Station_ID'] = Standardized_eDNA['Station_ID'].astype(str)
    pattern_Station_ID = r"^SS\d{5}$"
    Standardized_eDNA["Station_ID"] = Standardized_eDNA["Station_ID"].where(Standardized_eDNA["Station_ID"].str.match(pattern_Station_ID), '-')

if 'Sample_ID' in Standardized_eDNA.columns:
    Standardized_eDNA['Sample_ID'] = Standardized_eDNA['Sample_ID'].astype(str)
    pattern_Station_ID = r"^SAMP\d{4}$"
    Standardized_eDNA["Sample_ID"] = Standardized_eDNA["Sample_ID"].where(Standardized_eDNA["Sample_ID"].str.match(pattern_Station_ID), '-')

if 'Scientific_name' in Standardized_eDNA.columns:
    Standardized_eDNA['Scientific_name'] = Standardized_eDNA['Scientific_name'].astype(str)

if 'Sequence_ID' in Standardized_eDNA.columns:
    Standardized_eDNA['Sequence_ID'] = Standardized_eDNA['Sequence_ID'].astype(str)
    pattern_Station_ID = r"^SEQ\d{5}$"
    Standardized_eDNA["Sequence_ID"] = Standardized_eDNA["Sequence_ID"].where(Standardized_eDNA["Sequence_ID"].str.match(pattern_Station_ID), '-')

if 'eDNA_ReadCount' in Standardized_eDNA.columns:
    Standardized_eDNA['eDNA_ReadCount'] = pd.to_numeric(Standardized_eDNA["eDNA_ReadCount"], errors="coerce")
    min_eDNA_ReadCount = 1000
    max_eDNA_ReadCount = 25000
    Standardized_eDNA['eDNA_ReadCount'] = Standardized_eDNA['eDNA_ReadCount'].where((Standardized_eDNA['eDNA_ReadCount'] >= min_eDNA_ReadCount) & (Standardized_eDNA['eDNA_ReadCount'] <= max_eDNA_ReadCount), '-')

if 'Marker_Gene' in Standardized_eDNA.columns:
    Standardized_eDNA['Marker_Gene'] = Standardized_eDNA['Marker_Gene'].astype(str)

if 'Relative_Abundance_Percent' in Standardized_eDNA.columns:
    Standardized_eDNA['Relative_Abundance_Percent'] = pd.to_numeric(Standardized_eDNA["Relative_Abundance_Percent"], errors="coerce")
    min_Relative_Abundance_Percent = 0
    max_Relative_Abundance_Percent = 100
    Standardized_eDNA['Relative_Abundance_Percent'] = Standardized_eDNA['Relative_Abundance_Percent'].where((Standardized_eDNA['Relative_Abundance_Percent'] >= min_Relative_Abundance_Percent) & (Standardized_eDNA['Relative_Abundance_Percent'] <= max_Relative_Abundance_Percent), '-')

if 'eDNA_Collection_Date' in Standardized_eDNA.columns:
    Standardized_eDNA['eDNA_Collection_Date'] = pd.to_datetime(Standardized_eDNA['eDNA_Collection_Date'], errors='coerce')
    Standardized_eDNA['eDNA_Collection_Date'] = Standardized_eDNA['eDNA_Collection_Date'].dt.strftime('%d-%m-%Y')

Standard_otolith_col = ['Otolith_ID', 'Station_ID', 'Scientific_name', 'Image_File', 'Length_m', 'Width_m', 'Circularity', 'AspectRatio']

def normalize_column(col_name):
    norm = col_name.replace("_","").replace(" ","").replace("(","").replace(")","").lower()
    return norm

col_map = {}
for std_col in Standard_otolith_col:
    norm_std = normalize_column(std_col)
    for uploaded_col in Uploaded_data.columns:
        norm_uploaded = normalize_column(uploaded_col)
        if norm_std in norm_uploaded or norm_uploaded in norm_std:
            col_map[uploaded_col] = std_col

existing_cols = list(col_map.keys())
Standard_otolith_tab = Uploaded_data[existing_cols].copy()
Standard_otolith_tab.rename(columns=col_map, inplace=True)

Standard_otolith_tab = Standard_otolith_tab.drop_duplicates()

if 'Otolith_ID' in Standard_otolith_tab.columns:
    Standard_otolith_tab['Otolith_ID'] = Standard_otolith_tab['Otolith_ID'].astype(str)
    pattern_Station_ID = r"^CMLRE/OTL/\d{5}$"
    Standard_otolith_tab["Otolith_ID"] = Standard_otolith_tab["Otolith_ID"].where(Standard_otolith_tab["Otolith_ID"].str.match(pattern_Station_ID), '-')

if 'Station_ID' in Standard_otolith_tab.columns:
    Standard_otolith_tab['Station_ID'] = Standard_otolith_tab['Station_ID'].astype(str)
    pattern_Station_ID = r"^SS\d{5}$"
    Standard_otolith_tab["Station_ID"] = Standard_otolith_tab["Station_ID"].where(Standard_otolith_tab["Station_ID"].str.match(pattern_Station_ID), '-')

#we can also add species ID here

if 'Scientific_name' in Standard_otolith_tab.columns:
    Standard_otolith_tab['Scientific_name'] = Standard_otolith_tab['Scientific_name'].astype(str)

if 'Image_File' in Standard_otolith_tab.columns: #Yeh Modify krna baki hai abhi
    Standard_otolith_tab['Image_File'] = Standard_otolith_tab['Image_File'].astype(str) 

if 'Length_mm' in Standard_otolith_tab.columns: #Thinking to add a converter cause i think we'll need 
    Standard_otolith_tab['Length_mm'] = pd.to_numeric(Standard_otolith_tab["Length_mm"], errors="coerce")
    min_Length_mm = 0.006
    max_Length_mm = 25
    Standard_otolith_tab['Length_mm'] = Standard_otolith_tab['Length_mm'].where((Standard_otolith_tab['Length_mm'] >= min_Length_mm) & (Standard_otolith_tab['Length_mm'] <= max_Length_mm), '-')

if 'Width_mm' in Standard_otolith_tab.columns:
    Standard_otolith_tab['Width_mm'] = pd.to_numeric(Standard_otolith_tab["Width_mm"], errors="coerce")
    min_Length_mm = 0.006
    max_Length_mm = 25
    Standard_otolith_tab['Width_mm'] = Standard_otolith_tab['Width_mm'].where((Standard_otolith_tab['Width_mm'] >= min_Length_mm) & (Standard_otolith_tab['Width_mm'] <= max_Length_mm), '-')

if 'Circularity' in Standard_otolith_tab.columns:
    Standard_otolith_tab['Circularity'] = pd.to_numeric(Standard_otolith_tab["Circularity"], errors="coerce")
    min_Length_mm = 0.006 #Yet to be decided about the limit
    max_Length_mm = 25
    Standard_otolith_tab['Circularity'] = Standard_otolith_tab['Circularity'].where((Standard_otolith_tab['Circularity'] >= min_Length_mm) & (Standard_otolith_tab['Circularity'] <= max_Length_mm), '-')

if 'AspectRatio' in Standard_otolith_tab.columns:
    Standard_otolith_tab['AspectRatio'] = pd.to_numeric(Standard_otolith_tab["AspectRatio"], errors="coerce")
    min_Length_mm = 0.006 #Yet to be decided about the limit
    max_Length_mm = 25
    Standard_otolith_tab['AspectRatio'] = Standard_otolith_tab['AspectRatio'].where((Standard_otolith_tab['AspectRatio'] >= min_Length_mm) & (Standard_otolith_tab['AspectRatio'] <= max_Length_mm), '-')

Ultimate_Dataset = pd.concat([Standardized_format.reindex(columns=['Station_ID', 'Scientific_name', 'Locality', 'Temperature','Latitude', 'Longitude', 'Salinity/PSU', 'Dissolved_O2_mgL','Current_m_s', 'Collection_Date']),
                              Standard_taxonomy_tab.reindex(columns=['Station_ID', 'Species_ID', 'Scientific_name', 'Common_name','Phylum', 'Class', 'Order', 'Family', 'Genus', 'Species', 'Abundance']), 
                              Standardized_eDNA.reindex(columns=['Station_ID', 'Sample_ID', 'Scientific_name', 'Sequence_ID','eDNA_ReadCount', 'Marker_Gene', 'Relative_Abundance_Percent', 'eDNA_Collection_Date']),
                              Standard_otolith_tab.reindex(columns=['Otolith_ID', 'Station_ID', 'Scientific_name', 'Image_File','Length_m', 'Width_m', 'Circularity', 'AspectRatio'])], axis=1, sort=False)

print("Select the dataset by entering the corresponding number:")
print("1: Ultimate Dataset")
print("2: Standardized Format")
print("3: Taxonomy")
print("4: eDNA")
print("5: Otolith")

choice = int(input("Enter your choice (1-5): "))

if choice == 1:
    selected_data = Ultimate_Dataset
    print("You selected: Ultimate Dataset")

elif choice == 2:
    selected_data = Standardized_format
    print("You selected: Standardized Format")

elif choice == 3:
    selected_data = Standard_taxonomy_tab
    print("You selected: Taxonomy")

elif choice == 4:
    selected_data = Standardized_eDNA
    print("You selected: eDNA")

elif choice == 5:
    selected_data = Standard_otolith_tab
    print("You selected: Otolith")

else:
    selected_data = None
    print("Invalid choice! Please enter a number from 1 to 5.")

selected_data
```
